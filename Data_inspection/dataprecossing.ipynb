```python
df2["salary_available"] = df2["placement_status"].astype(int)
df2["salary_package_lpa"] = df2["salary_package_lpa"].where(
    df["placement_status"] == 1, other=float("nan")
)
df2
```

```python
from sklearn.preprocessing import MinMaxScaler
tech_cols = [
    "coding_skills",
    "dsa_score",
    "ml_knowledge",
    "system_design",
    "aptitude_score",
    "certifications"
]


scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df2[tech_cols])

df_scaled = pd.DataFrame(df_scaled, columns=tech_cols)

df2["tech_skill"] = df_scaled.mean(axis=1)

```

```python
soft_cols = ['communication_skills','extracurriculars']
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df2[soft_cols])

df_scaled = pd.DataFrame(df_scaled, columns=soft_cols)

df2["soft_skill"] = df_scaled.mean(axis=1)

```

```python
practical_cols = ['internships','open_source_contributions','hackathons','projects_count']
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df2[practical_cols])

df_scaled = pd.DataFrame(df_scaled, columns=practical_cols)

df2["practical_skill"] = df_scaled.mean(axis=1)

```

```python
acad_cols = ["cgpa", "backlogs"]

scaler = MinMaxScaler()
scaled = scaler.fit_transform(df2[acad_cols])

df2["cgpa_scaled"] = scaled[:, 0]
df2["backlogs_scaled"] = scaled[:, 1]
df2["academic_points"] = (
    0.7 * df2["cgpa_scaled"] +
    0.3 * (1 - df2["backlogs_scaled"])
)
```

```python
df2
```

```python
tier_map = {'Tier-1': 1, 'Tier-2': 2, 'Tier-3': 3}
df2['college_tier'] = df2['college_tier'].map(tier_map)

# Verify
print(df2['college_tier'].value_counts())
```

```python
df2 = pd.get_dummies(df2, columns=['branch'], drop_first=True)


branch_cols = df2.filter(like='branch').columns
df2[branch_cols] = df2[branch_cols].astype(int)
```

```python
raw_cols = [
    'coding_skills', 'dsa_score', 'ml_knowledge',
    'system_design', 'aptitude_score', 'certifications',
    'communication_skills', 'extracurriculars',
    'internships', 'projects_count', 'hackathons',
    'open_source_contributions', 'cgpa', 'backlogs'
]
df2 = df2.drop(columns=raw_cols)

print("Remaining columns:", df2.columns.tolist())
print("Shape:", df2.shape)
```
